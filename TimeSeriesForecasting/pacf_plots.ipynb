# Checking PACF plot for Autoregressors

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
from statsmodels.tsa.ar_model import AutoReg
```

```python
df = pd.read_csv("data_3_clean.csv", parse_dates=True, index_col=0)
df.head()
```

```python
print(f"shape of dataset = {df.shape}")

d = df[['Reference']]
print(d.shape)
```

```python
d.plot()
```

```python
from statsmodels.tsa.stattools import adfuller

dtest = adfuller(d['Reference'], autolag="AIC")

print("1. ADF :", dtest[0])
print("2. P-value :", dtest[1])
print("3. Num of lags : ", dtest[3])
```
