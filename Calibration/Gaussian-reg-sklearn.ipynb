```python
import numpy as np
from sklearn.gaussian_process.kernels import DotProduct, WhiteKernel
from sklearn.gaussian_process import GaussianProcessRegressor

combined_predictions = np.array([])
combined_actuals = np.array([])

for i in range(52):
    X = data.loc[:, data.columns.drop(column_groups[i])]
    y = data[column_groups[i][0]]
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=32, test_size=0.2, shuffle=True)

    kernel = DotProduct() + WhiteKernel()
    regressor = GaussianProcessRegressor(kernel=kernel)
    regressor.fit(X_train, y_train)

    predictions = regressor.predict(X_test)
    combined_predictions = np.concatenate((combined_predictions, predictions))
    combined_actuals = np.concatenate((combined_actuals, y_test))

overall_r2 = 1 - (np.sum((combined_actuals - combined_predictions) ** 2) / np.sum((combined_actuals - np.mean(combined_actuals)) ** 2))
print("Overall R-squared Score:", overall_r2)
```
