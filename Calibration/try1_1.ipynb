# Training the model on the complete data using Gradient Boosting Regression

```python
import numpy as np
import pandas as pd 
import os 
import sys

from sklearn import model_selection
from sklearn import metrics
from sklearn import preprocessing
from sklearn import ensemble

from tqdm.notebook import tqdm
import pickle

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
final_dict_save_path = './data/final_dict.pkl'
pm25_index_to_name_path = './data/p25_index_to_name.pkl'
p25_name_to_index_path = './data/p25_name_to_index.pkl'

with open(final_dict_save_path, 'rb') as f:
    final_dict = pickle.load(f)

with open(pm25_index_to_name_path, 'rb') as f:
    pm25_name_dict = pickle.load(f)

with open(p25_name_to_index_path, 'rb') as f:
    pm25_name_to_index = pickle.load(f)
```

- i want to use 13 sensors for training and the remaining 1 for testing
- for each sensor we have entries for 92 time stamps 
- 92 * 1 * 13 will be for testing = 1196
- 92 * 13 * 13 will be used for training = 14352

### Let's first make the training and testing data


```python
param_grid = {
    'learning_rate': [0.01, 0.05, 0.1, 0.5],
    'n_estimators': [50, 100, 200, 300, 400, 500],
    'max_depth': [3, 4, 5],
    'min_samples_split': [5],
    "loss": ['squared_error']
    }

gbr = ensemble.GradientBoostingRegressor()
gbr_cv = model_selection.GridSearchCV(gbr, param_grid=param_grid, cv=5)
```

```python
np.random.RandomState(482)
test_sensor_index = np.random.randint(0, len(pm25_name_dict))
test_sensor_name = pm25_name_dict[test_sensor_index]

print(f"test_sensor_index = {test_sensor_index}, test_sensor_name = {test_sensor_name}")



for sensor_index, sensor_name in enumerate(final_dict.keys()):
    # print('----------------------------------------------')
    # print(f"sensor_index = {sensor_index}, sensor_name = {sensor_name}\n")

    if sensor_name == test_sensor_name:
        continue # skipping the test sensor for now

    num_rows, num_columns = final_dict[test_sensor_name]['2023-01-05 13:00:00']['X'].shape
    print(f"num_rows = {num_rows}, num_columns = {num_columns}")
    X_train, X_test = np.empty((0, num_columns), dtype=np.float32), np.empty((0, num_columns), dtype=np.float32)
    y_train, y_test = np.empty((0, 1), dtype=np.float32), np.empty((0, 1), dtype=np.float32)

    for time in final_dict[sensor_name].keys():

        # looping over time

        Xtrain = final_dict[sensor_name][time]['X'].loc[final_dict[sensor_name][time]['X'].index != test_sensor_name].values
        Xtest = final_dict[sensor_name][time]['X'].loc[test_sensor_name].values.reshape(1, -1)
        ytrain = final_dict[sensor_name][time]['y'].loc[final_dict[sensor_name][time]['y'].index != test_sensor_index].values.reshape(-1, 1)
        ytest = final_dict[sensor_name][time]['y'].iloc[test_sensor_index].values.reshape(1, -1)

        # print(f"Xtrain.shape = {Xtrain.shape}, ytrain.shape = {ytrain.shape}")
        # print(f"Xtest.shape = {Xtest.shape}, ytest.shape = {ytest.shape}")
        
        X_train = np.concatenate((X_train, Xtrain), axis=0)
        X_test = np.concatenate((X_test, Xtest), axis=0)
        y_train = np.concatenate((y_train, ytrain), axis=0)
        y_test = np.concatenate((y_test, ytest), axis=0)

    scaler = preprocessing.StandardScaler()

    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    print(f"X_train.shape = {X_train.shape}, y_train.shape = {y_train.shape}")
    print(f"X_test.shape = {X_test.shape}, y_test.shape = {y_test.shape}")

    gbr_cv.fit(X_train, y_train)

    print('----------------- Best Params:')
    print(gbr_cv.best_params_)

    print('----------------- Training Scores : ')

    print(f"train r2 score = {gbr_cv.score(X_train, y_train)}")

    print('----------------- Testing Scores : ')

    gbr_best = gbr_cv.best_estimator_

    y_pred = gbr_best.predict(X_test)
    print(f"y_pred.shape = {y_pred.shape}")

    print(f"r2 score = {metrics.r2_score(y_test, y_pred)}")
    print(f"mse = {metrics.mean_squared_error(y_test, y_pred)}")

    print('----------------- Prediction Graph : ')

    # let's plot the predictions and the true values

    x = np.arange(len(y_pred))
    plt.figure()
    plt.plot(x, y_test.reshape(-1), color='blue', label='true')
    plt.plot(x, y_pred, color='red', label='predictions')
    plt.title(f"GradientBoostRegressor, n_estimators={gbr_best.get_params()['n_estimators']}, lr={gbr_best.get_params()['learning_rate']}")
    plt.grid()
    plt.legend()
    plt.show()
```

- now we have the trained model, let's try to run inference on the test_sensor_data that we didn't used for training

```python
gbr_best = gbr_cv.best_estimator_
gbr_best
```

```python
print(f"test_sensor_name = {test_sensor_name}")
# final_dict[test_sensor_name]

# let's for each time stamp store a single prediction : easy for plotting

y_true_plot = []
y_pred_plot = []

y_true = np.empty((0, 1), dtype=np.float32)
y_model_pred = np.empty((0, 1), dtype=np.float32)

for time in final_dict[test_sensor_name]:

    # 92 time stamps

    X_test = final_dict[test_sensor_name][time]['X']
    y_test = final_dict[test_sensor_name][time]['y']

    y_pred = gbr_best.predict(X_test)

    y_true = np.concatenate((y_true, y_test), axis=0)
    y_model_pred = np.concatenate((y_model_pred, y_pred.reshape(-1, 1)), axis=0)

    y_true_plot.append(y_test.values[0][0])
    # random_index = np.random.randint(0, len(y_pred))
    y_pred_plot.append(y_pred[5])


print(f"y_true.shape = {y_true.shape}, y_model_pred.shape = {y_model_pred.shape}")
print(len(y_true_plot), len(y_pred_plot))    

print(f"r2 score = {metrics.r2_score(y_true, y_model_pred)}")
print(f"mse = {metrics.mean_squared_error(y_true, y_model_pred)}")

# let's plot the predictions and the true values

x = np.arange(len(y_true_plot))
plt.figure()
plt.plot(x, np.array(y_true_plot), color='blue', label='true')
plt.plot(x, np.array(y_pred_plot), color='red', label='predictions')
plt.title(f"GradientBoostRegressor, n_estimators={gbr_best.get_params()['n_estimators']}, lr={gbr_best.get_params()['learning_rate']}")
plt.grid()
plt.legend()
plt.show()
```
